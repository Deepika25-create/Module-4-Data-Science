```python
#Sorting with custom order
from pandas.api.types import CategoricalDtype
grades=['A','B','C','D','E']
grade_dtype=CategoricalDtype(categories=grades,ordered=True)

data={
    "Name":['Kashish','Simran','Radha'],
    "Grade":['B','A','C']
}
df_grades=pd.DataFrame(data)
df_grades
```

```python
#Convert Grade to Categorical with custom order
df_grades['Grade']=df_grades['Grade'].astype(grade_dtype)
```

```python
#Sort by age
sort_age=df_grades.sort_values(by='Grade')
sort_age
```

```python
#Sorting by Index level in MultiIndex
arrays=[['Team A','Team B','Team B','Team A'],['Player 1', 'Player 2','Player 2','Player 1']]
index=pd.MultiIndex.from_arrays(arrays,names=('Team','Player'))

data={'Score':[78,90,78,92]}
multi_df=pd.DataFrame(data,index=index)

multi_df
```

```python
#Sorting by Player level
sort_multi=multi_df.sort_index(level='Player')
sort_multi
```

```python
#GROUPBY
import pandas as pd
data={
    'Course':['AI/ML','AI/ML','FED','FED','Data Science','Data Science'],
    'Resourceperson':['Ananya','Balram','Ananyae','Balram','Ananya','Balram'],
    'Revenue':[1600,1500,700,1200,1600,1800]
}
df=pd.DataFrame(data)
df
```

```python
#Grouping data by Course and calculating revenue per person
grouped=df.groupby('Course')['Revenue'].sum()
grouped
```

```python
#Grouping data by multiple column and calculating revenue
grouped=df.groupby(['Course','Resourceperson'])['Revenue'].sum()
grouped
```

```python
mean=df.groupby('Course')['Revenue'].agg(['sum','mean'])
mean
```

```python
#Adding new column with percentage revenue
df['Revenue_Percentage']=df.groupby('Course').transform(lambda x:(x/x.sum()*100))
df
```

```python
#Filtering Group based on Condition
filtered=df.groupby('Course').filter(lambda x:x['Revenue'].sum()>2000)
filtered
```
